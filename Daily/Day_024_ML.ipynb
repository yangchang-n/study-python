```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('./auto-mpg.csv', header = None)
df.head()
```

```python
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'accelaration', 'model year', 'origin', 'name']
df.head()
```

```python
pd.set_option('display.max_columns', 10)
print(df.head())
```

```python
print(df.info())
print()
print(df.describe())
```

```python
print(df['horsepower'].unique())
print()
df['horsepower'].replace('?', np.nan, inplace = True)
df.dropna(subset = ['horsepower'], axis = 0, inplace = True)
df['horsepower'] = df['horsepower'].astype('float')
print(df.info())
```

```python
ndf = df[['mpg', 'cylinders', 'horsepower', 'weight']]
print(ndf.head())
```

```python
ndf.plot(kind = 'scatter', x = 'weight', y = 'mpg', c = 'coral', s = 30, figsize = (16, 8))
plt.show()
plt.close()
```

```python
fig = plt.figure(figsize = (16, 8))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
sns.regplot(x = 'weight', y = 'mpg', data = ndf, ax = ax1, color = 'coral')
sns.regplot(x = 'weight', y = 'mpg', data = ndf, ax = ax2, fit_reg = False, color = 'coral')
plt.show()
plt.close()
```

```python
sns.jointplot(x = 'weight', y = 'mpg', data = ndf)
sns.jointplot(x = 'weight', y = 'mpg', kind = 'reg', data = ndf)
plt.show()
```

```python
grid_ndf = sns.pairplot(ndf)
plt.show()
plt.close()
```

```python
X = ndf[['weight']]
y = ndf['mpg']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)

print('train data # :', len(X_train))
print('test data # :', len(X_test))
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)

r_square = lr.score(X_test, y_test)
print(r_square)
```

```python
print(lr.coef_, lr.intercept_)
```

```python
y_hat = lr.predict(X)
plt.figure(figsize = (16, 8))
ax1 = sns.kdeplot(y, label = 'y')
ax2 = sns.kdeplot(y_hat, label = 'y_hat', ax = ax1)
plt.legend()
plt.show()
```

```python
print(X_train.shape, X_test.shape)
```

```python
y_hat_2 = pr.predict(X_test_poly)
fig = plt.figure(figsize = (16, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(X_train, y_train, 'o', label = 'Train Data')
ax.plot(X_test, y_hat_2, 'r+', label = 'Predicted Value')
ax.legend(loc = 'best')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
plt.close()
```

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
df.head()
```

```python
print(df.info())
print()
rdf = df.drop(['deck', 'embark_town'], axis = 1)
print(rdf.columns.values)
print()
rdf = rdf.dropna(subset = ['age'], how = 'any', axis = 0)
print(len(rdf))
print()
most_freq = rdf['embarked'].value_counts(dropna = True).idxmax()
print(most_freq)
print()
print(rdf.describe(include = 'all'))
print()
rdf['embarked'].fillna(most_freq, inplace = True)
```

```python
ndf = rdf[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'embarked']]
print(ndf.head())
print()
onehot_sex = pd.get_dummies(ndf['sex'])
ndf = pd.concat([ndf, onehot_sex], axis = 1)
onehot_embarked = pd.get_dummies(ndf['embarked'], prefix = 'town')
ndf = pd.concat([ndf, onehot_embarked], axis = 1)
ndf.drop(['sex', 'embarked'], axis = 1, inplace = True)
print(ndf.head())
```

```python
X = ndf[['pclass', 'age', 'sibsp', 'parch', 'female', 'male', 'town_C', 'town_Q', 'town_S']]
y = ndf['survived']

from sklearn import preprocessing

X = preprocessing.StandardScaler().fit(X).transform(X)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
print(X_train.shape, y_train.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)

y_hat = knn.predict(X_test)
aa = knn.predict_proba(X_test)

print(y_hat[ : 10])
print(y_test.values[ : 10])

from sklearn import metrics

knn_matrix = metrics.confusion_matrix(y_test, y_hat)
print(knn_matrix)

knn_report = metrics.classification_report(y_test, y_hat)
print(knn_report)
```

```python
from sklearn import svm

svm_model = svm.SVC(kernel = 'rbf')
svm_model.fit(X_train, y_train)
y_hat_2 = svm_model.predict(X_test)

print(y_hat_2[ : 10])
print(y_test.values[ : 10])

svm_matrix = metrics.confusion_matrix(y_test, y_hat_2)
print(svm_matrix)

svm_report = metrics.classification_report(y_test, y_hat_2)
print(svm_report)
```

```python
import pandas as pd
import numpy as np

uci_path = 'https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data'
df = pd.read_csv(uci_path, header = None)
df.columns = ['id', 'clump', 'cell_size', 'cell_shape', 'adhesion', 'epithlial', 'bare_nuclei', 'chromatin', 'normal_nucleoli', 'mitoses', 'class']
```

```python
df.head()
```

```python
print(df['bare_nuclei'].unique())
```

```python
df.info()
```

```python
df['bare_nuclei'].replace('?', np.nan, inplace = True)
df.dropna(subset = ['bare_nuclei'], axis = 0, inplace = True)
df['bare_nuclei'] = df['bare_nuclei'].astype('int')
df.info()
```

```python
feature_names = ['clump', 'cell_size', 'cell_shape', 'adhesion', 'epithlial', 'bare_nuclei', 'chromatin', 'normal_nucleoli', 'mitoses']
X = df[feature_names]
y = df['class']

from sklearn import preprocessing

X = preprocessing.StandardScaler().fit(X).transform(X)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
print(X_train.shape, y_train.shape)
```

```python
from sklearn import tree

tree_model = tree.DecisionTreeClassifier(criterion = 'entropy', max_depth = 5)
tree_model.fit(X_train, y_train)
y_hat_3 = tree_model.predict(X_test)

print(y_hat_3[ : 10])
print(y_test.values[ : 10])

tree_matrix = metrics.confusion_matrix(y_test, y_hat_3)
print(tree_matrix)

tree_report = metrics.classification_report(y_test, y_hat_3)
print(tree_report)
```

```python
plt.figure(figsize = (20, 15))

tree.plot_tree(tree_model,
               class_names = 'class',
               feature_names = feature_names,
               impurity = True, filled = True,
               rounded = True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

uci_path = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv'
df = pd.read_csv(uci_path, header = 0)
```

```python
print(df.head())
print()
print(df.info())
print()
print(df.describe())
```

```python
X = df.iloc[ : , : ]
print(X[ : 5])
print()

from sklearn import preprocessing

X = preprocessing.StandardScaler().fit(X).transform(X)

print(X[ : 5])
```

```python
from sklearn import cluster

kmeans = cluster.KMeans(init = 'k-means++', n_clusters = 5, n_init = 10)
kmeans.fit(X)
cluster_label = kmeans.labels_
print(cluster_label)
print()
df['Cluster'] = cluster_label
print(df.head())
```

```python
df.plot(kind = 'scatter', x = 'Grocery', y = 'Frozen', c = 'Cluster', cmap = 'Set1', colorbar = False, figsize = (8, 8))
df.plot(kind = 'scatter', x = 'Milk', y = 'Delicassen', c = 'Cluster', cmap = 'Set1', colorbar = False, figsize = (8, 8))
plt.show()
plt.close()
```
