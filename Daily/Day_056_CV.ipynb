```python
import cv2
from google.colab.patches import cv2_imshow

src = cv2.imread('iu_sample.jpg')
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
binary = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 467, 35)

cv2_imshow(binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
