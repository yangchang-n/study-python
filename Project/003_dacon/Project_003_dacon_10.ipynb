```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
val = pd.read_csv('/content/drive/MyDrive/Project_003/val.csv')
```

```python
columns = []

for i in range(1, 30 + 1) :
  columns.append('V{}'.format(i))

print(columns)
```

```python
val_col = val.copy()
val_col = val_col[columns]
val_col.head()
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
```

```python
val_ss = pd.DataFrame(ss.fit_transform(val_col))
val_ss.columns = columns
val_ss.head()
```

```python
val_ss = pd.concat([val_ss, val['Class']], axis = 1)
val_ss.head()
```

```python
val_good = val_ss.copy()
val_good = val_good[val_good['Class'] == 0]
print(len(val_good))
```

```python
val_bad = val_ss.copy()
val_bad = val_bad[val_bad['Class'] == 1]
print(len(val_bad))
```

```python
val_good['V1'].mean()
```

```python
val_priority ={}

for i in range(1, 30 + 1) :
  val_priority['V{}'.format(i)] = abs(val_good['V{}'.format(i)].mean() - val_bad['V{}'.format(i)].mean())

print(val_priority)
```

```python
val_priority_list = sorted(val_priority.items(), key = lambda x : x[1], reverse = True)
print(val_priority_list)
```

```python
for i in range(10) :
  print('{}. {}'.format(i + 1, val_priority_list[i][0]))
```
