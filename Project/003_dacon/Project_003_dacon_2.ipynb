```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('/content/drive/MyDrive/Project_003/train.csv')
```

```python
train.head()
```

```python
train_noid = train.copy()
train_noid.head()
```

```python
train_noid.drop(['ID'], axis = 1, inplace = True)
```

```python
train_noid.head()
```

```python
from sklearn import preprocessing as pp

ss = pp.StandardScaler()
```

```python
train_scaled = ss.fit_transform(train_noid)
train_scaled = pd.DataFrame(train_scaled)
train_scaled.head()
```

```python
columns = []
for i in range(1, 30 + 1) :
  columns.append('V{}'.format(i))
print(columns)
```

```python
train_scaled.columns = columns
train_scaled.head()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

plt.figure(figsize = (18, 18))
sb.heatmap(data = train_scaled.corr(), annot = True, fmt = '.2f', cmap = 'Blues')
plt.show()
```

```python
# V29, V30 ??
# V7, V20 with V29 == + corr
# V2, V5 with V29 == - corr
# V3 with V30 == - corr
```

```python
train_select = train_scaled[['V2', 'V3', 'V5', 'V7', 'V20', 'V29', 'V30']]
```

```python
train_select.head()
```

```python
plt.figure(figsize = (18, 18))
sb.heatmap(data = train_select.corr(), annot = True, fmt = '.2f', cmap = 'Blues')
plt.show()
```

```python
train_select.describe()
```

```python
train_select_2 = train_select.copy()
train_select_2 = train_select_2.iloc[ : 10000]
train_select_2.shape
```

```python
from sklearn.cluster import DBSCAN

model_db = DBSCAN(eps = 5, min_samples = 10)
predict = pd.DataFrame(model_db.fit_predict(train_select_2))
predict.columns = ['predict']
visual = pd.concat([train_select_2, predict], axis = 1)
print(visual)
```

```python
visual[visual['predict'] == -1].count()
```

```python
len(predict.values)
```

```python
color = []

for i in range(len(predict.values)) :
  if predict.values[i][0] == -1 :
    color.append('r')
  else :
    color.append('b')

print(color.count('r'))
```

```python
train_select_2.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
```

```python
model_pipe = make_pipeline(PCA())
model_pipe.fit(train_select_2)
```

```python
model = PCA(n_components = 3)
train_pca = model.fit_transform(train_select_2)
train_pca.shape
```

```python
fig = plt.figure(figsize = (15, 15))
ax = fig.gca(projection = '3d')
ax.scatter(train_pca[ : , 0], train_pca[ : , 1], train_pca[ : , 2], alpha = 0.5, c = color)
plt.show()
```

```python
val = pd.read_csv('/content/drive/MyDrive/Project_003/val.csv')
```

```python
val.shape
```

```python
val_noid = val.copy()
val_noid.drop(['ID', 'Class'], axis = 1, inplace = True)
val_noid.head()
```

```python
val_scaled = ss.fit_transform(val_noid)
val_scaled = pd.DataFrame(val_scaled)
val_scaled.head()
```

```python
val_scaled.columns = columns
val_scaled.head()
```

```python
val_select = val_scaled[['V2', 'V3', 'V5', 'V7', 'V20', 'V29', 'V30']]
val_select.head()
```

```python
model_pipe_val = make_pipeline(PCA())
model_pipe_val.fit(val_select)
```

```python
model_val = PCA(n_components = 3)
val_pca = model_val.fit_transform(val_select)
val_pca.shape
```

```python
model_db = DBSCAN(eps = 5, min_samples = 10)
predict_val = pd.DataFrame(model_db.fit_predict(val_select))
predict_val.columns = ['predict_val']
visual_val = pd.concat([val_select, predict_val], axis = 1)
print(visual_val)
```

```python
visual_val[visual_val['predict_val'] == -1].count()
```

```python
color_val = []

for i in range(len(predict_val.values)) :
  if predict_val.values[i][0] == -1 :
    color_val.append('r')
  else :
    color_val.append('b')

print(color_val.count('r'))
```

```python
val['Class'].values
```

```python
color_ans = []

for i in range(len(val.values)) :
  if val['Class'].values[i] == 1 :
    color_ans.append('r')
  else :
    color_ans.append('b')

print(color_ans.count('r'))
```

```python
fig = plt.figure(figsize = (15, 15))
ax = fig.gca(projection = '3d')
ax.scatter(val_pca[ : , 0], val_pca[ : , 1], val_pca[ : , 2], alpha = 0.5, c = color_val)
plt.show()
```

```python
fig = plt.figure(figsize = (15, 15))
ax = fig.gca(projection = '3d')
ax.scatter(val_pca[ : , 0], val_pca[ : , 1], val_pca[ : , 2], alpha = 0.5, c = color_ans)
plt.show()
```
