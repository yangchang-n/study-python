```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
val_scaled = ss.fit_transform(val_feature)
val_scaled = pd.DataFrame(val_scaled)

columns = []
for i in range(1, 30 + 1) :
  columns.append('V{}'.format(i))

val_scaled.columns = columns

val_scaled.head()
```

```python
val_scaled_onlyhighcorr = val_scaled.copy()
val_scaled_onlyhighcorr = val_scaled_onlyhighcorr[['V1', 'V2', 'V3', 'V5', 'V7', 'V9', 'V10']]
val_scaled_onlyhighcorr.head()
```

```python
val_arr = val_scaled_onlyhighcorr.to_numpy()
val_arr[ : 2]
```

```python
val_scaled_onlyhighcorr.describe()
```

```python
answer = []
min = 0.9988
max = 1.0012

for i in range(len(val_arr)) :
  if ((val_arr[i][0] >= (val_arr[i][4] * min)) and (val_arr[i][0] <= (val_arr[i][4] * max))) or\
  ((val_arr[i][2] >= (val_arr[i][4] * min)) and (val_arr[i][2] <= (val_arr[i][4] * max))) or\
  ((val_arr[i][3] >= (val_arr[i][4] * min)) and (val_arr[i][3] <= (val_arr[i][4] * max))) or\
  ((val_arr[i][5] >= (val_arr[i][4] * min)) and (val_arr[i][5] <= (val_arr[i][4] * max))) or\
  ((val_arr[i][6] >= (val_arr[i][4] * min)) and (val_arr[i][6] <= (val_arr[i][4] * max))) :
    answer.append(1)
  else :
    answer.append(0)

answer.count(1)
```

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
```

```python
model_pipe = make_pipeline(PCA())
model_pipe.fit(val_feature)
```

```python
model = PCA(n_components = 3)
val_pca = model.fit_transform(val_feature)
val_pca.shape
```

```python
color_ans = []

for i in range(len(val.values)) :
  if val['Class'].values[i] == 1 :
    color_ans.append('r')
  else :
    color_ans.append('b')

print(color_ans.count('r'))
```

```python
color_pre = []

for i in range(len(answer)) :
  if answer[i] == 1 :
    color_pre.append('r')
  else :
    color_pre.append('b')

print(color_pre.count('r'))
```

```python
fig = plt.figure(figsize = (15, 15))
ax = fig.gca(projection = '3d')
ax.scatter(val_pca[ : , 0], val_pca[ : , 1], val_pca[ : , 2], alpha = 0.5, c = color_ans)
plt.show()
```
