```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('/content/drive/MyDrive/Project_003/train.csv')
```

```python
train.head()
```

```python
train_noid = train.copy()
train_noid.head()
```

```python
train_noid.drop(['ID'], axis = 1, inplace = True)
```

```python
train_noid.head()
```

```python
from sklearn import preprocessing as pp

ss = pp.StandardScaler()
```

```python
train_scaled = ss.fit_transform(train_noid)
train_scaled = pd.DataFrame(train_scaled)
train_scaled.head()
```

```python
columns = []
for i in range(1, 30 + 1) :
  columns.append('V{}'.format(i))
print(columns)
```

```python
train_scaled.columns = columns
train_scaled.head()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

plt.figure(figsize = (18, 18))
sb.heatmap(data = train_scaled.corr(), annot = True, fmt = '.2f', cmap = 'Blues')
plt.show()
```

```python
# V29, V30 ??
# V7, V20 with V29 == + corr
# V2, V5 with V29 == - corr
# V3 with V30 == - corr
```

```python
train_select = train_scaled[['V2', 'V3', 'V5', 'V7', 'V20', 'V29', 'V30']]
```

```python
train_select.head()
```

```python
plt.figure(figsize = (18, 18))
sb.heatmap(data = train_select.corr(), annot = True, fmt = '.2f', cmap = 'Blues')
plt.show()
```

```python
train_select.describe()
```

```python
train_select_2 = train_select.copy()
train_select_2 = train_select_2.iloc[ : 10000]
train_select_2.shape
```

```python
from sklearn.cluster import DBSCAN

model_db = DBSCAN(eps = 5, min_samples = 10)
predict = pd.DataFrame(model_db.fit_predict(train_select_2))
predict.columns = ['predict']
visual = pd.concat([train_select_2, predict], axis = 1)
print(visual)
```

```python
visual[visual['predict'] == -1].count()
```

```python
len(predict.values)
```

```python
color = []

for i in range(len(predict.values)) :
  if predict.values[i][0] == -1 :
    color.append('r')
  else :
    color.append('b')

print(color.count('r'))
```

```python
train_select_2.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
```

```python
model_pipe = make_pipeline(PCA())
model_pipe.fit(train_select_2)
```

```python
model = PCA(n_components = 3)
train_pca = model.fit_transform(train_select_2)
train_pca.shape
```

```python
fig = plt.figure(figsize = (15, 15))
ax = fig.gca(projection = '3d')
ax.scatter(train_pca[ : , 0], train_pca[ : , 1], train_pca[ : , 2], alpha = 0.5, c = color)
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
test = pd.read_csv('/content/drive/MyDrive/Project_003/test.csv')
```

```python
test.head()
```

```python
test_noid = test.copy()
test_noid.drop(['ID'], axis = 1, inplace = True)
test_noid.head()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

plt.figure(figsize = (18, 18))
sb.heatmap(data = test_noid.corr(), annot = True, fmt = '.2f', cmap = 'Blues')
plt.show()
```

```python
test_select = test_noid[['V2', 'V3', 'V5', 'V7', 'V20', 'V29', 'V30']]
test_select.head()
```

```python
test_select.shape
```

```python
test_select_1 = test_select.iloc[0 : 25000 + 1]
test_select_2 = test_select.iloc[25001 : 50000 + 1]
test_select_3 = test_select.iloc[50001 : 75000 + 1]
test_select_4 = test_select.iloc[75001 : 100000 + 1]
test_select_5 = test_select.iloc[100001 : 125000 + 1]
test_select_6 = test_select.iloc[125001 : ]
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps = 6, min_samples = 10)
predict_1 = pd.DataFrame(model.fit_predict(test_select_1))
predict_1.columns = ['predict']
predict_1[predict_1['predict'] == -1].count()
```

```python
predict_2 = pd.DataFrame(model.fit_predict(test_select_2))
predict_2.columns = ['predict']
predict_2[predict_2['predict'] == -1].count()
```

```python
predict_3 = pd.DataFrame(model.fit_predict(test_select_3))
predict_3.columns = ['predict']
predict_3[predict_3['predict'] == -1].count()
```

```python
predict_4 = pd.DataFrame(model.fit_predict(test_select_4))
predict_4.columns = ['predict']
predict_4[predict_4['predict'] == -1].count()
```

```python
predict_5 = pd.DataFrame(model.fit_predict(test_select_5))
predict_5.columns = ['predict']
predict_5[predict_5['predict'] == -1].count()
```

```python
predict_6 = pd.DataFrame(model.fit_predict(test_select_6))
predict_6.columns = ['predict']
predict_6[predict_6['predict'] == -1].count()
```

```python
total_predict = pd.concat([predict_1, predict_2, predict_3, predict_4, predict_5, predict_6], ignore_index = True)
total_predict.head()
```

```python
total_predict.shape
```

```python
total_predict['predict']
```

```python
result = []

for i in range(len(total_predict)) :
  if total_predict['predict'][i] == -1 :
    result.append(1)
  else :
    result.append(0)

result.count(1)
```

```python
result_color = []

for i in range(len(result)) :
  if result[i] == 1 :
    result_color.append('r')
  else :
    result_color.append('b')

result_color.count('r')
```

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
```

```python
model_pipe = make_pipeline(PCA())
model_pipe.fit(test_noid)
```

```python
model = PCA(n_components = 3)
test_pca = model.fit_transform(test_noid)
test_pca
```

```python
fig = plt.figure(figsize = (15, 15))
ax = fig.gca(projection = '3d')
ax.scatter(test_pca[ : , 0], test_pca[ : , 1], test_pca[ : , 2], alpha = 0.5, c = result_color)
plt.show()
```

```python
pd_result = pd.DataFrame(result)
pd_result.columns = ['Class']
pd_result.head()
```

```python
pd_result.shape
```

```python
answer = pd.concat([test['ID'], pd_result], axis = 1)
answer.head()
```
