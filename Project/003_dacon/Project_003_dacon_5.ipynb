```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
val = pd.read_csv('/content/drive/MyDrive/Project_003/val.csv')
```

```python
val.head()
```

```python
color_ans = []

for i in range(len(val)) :
  if val['Class'][i] == 1 :
    color_ans.append('r')
  else :
    color_ans.append('b')

color_ans.count('r')
```

```python
fig = plt.figure(figsize = (12, 12))
plt.scatter(val['V12'], val['V17'], alpha = 0.5, c = color_ans)
plt.plot([-15, 5], [10, -10]) # y = - x - 5
plt.show()
```

```python
val_pred = []
color_pred = []

for i in range(len(val)) :
  if val.iloc[i]['V17'] >= - val.iloc[i]['V12'] - 5 :
    val_pred.append(0)
    color_pred.append('b')
  else :
    val_pred.append(1)
    color_pred.append('r')

print(val_pred.count(1), val_pred.count(0))
```

```python
fig = plt.figure(figsize = (12, 12))
plt.scatter(val['V12'], val['V17'], alpha = 0.5, c = color_pred)
plt.plot([-15, 5], [10, -10]) # y = - x - 5
plt.show()
```

```python
test = pd.read_csv('/content/drive/MyDrive/Project_003/test.csv')
```

```python
test.head()
```

```python
fig = plt.figure(figsize = (12, 12))
plt.scatter(test['V12'], test['V17'], alpha = 0.5, c = 'b')
plt.plot([-15, 5], [8.8, -11.2]) # y = - x - 6.2
plt.show()
```

```python
test_pred = []
test_color = []

for i in range(len(test)) :
  if test.iloc[i]['V17'] >= - test.iloc[i]['V12'] - 6.2 :
    test_pred.append(0)
    test_color.append('b')
  else :
    test_pred.append(1)
    test_color.append('r')

print(test_pred.count(1), test_pred.count(0))
```

```python
fig = plt.figure(figsize = (12, 12))
plt.scatter(test['V12'], test['V17'], alpha = 0.5, c = test_color)
plt.plot([-15, 5], [8.8, -11.2]) # y = - x - 6.2
plt.show()
```

```python
test_pred_pd = pd.DataFrame(test_pred)
test_pred_pd.columns = ['Class']
test_pred_pd.shape
```

```python
answer = pd.concat([test['ID'], test_pred_pd], axis = 1)
answer.set_index('ID', inplace = True)
answer.head()
```
